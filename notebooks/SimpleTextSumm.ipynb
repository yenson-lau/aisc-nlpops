```python
import nltk
import re
from urllib import request

url = "http://www.gutenberg.org/cache/epub/10/pg10.txt"
chapter = 1
text = []
write = False
with request.urlopen(url) as response:
  for line in response.read().decode("utf-8").splitlines():
    if line != '':
      if not write and line.split()[0].split(':')[0]==str(chapter):
        write = True
      if write and line.split()[0].split(':')[0]==str(chapter+1):
        break
      if write:
        text.append(re.sub(r'\d+:\d+ ', '', line))
  text = '\n'.join(text)
  
print(text)
```
